#### 2. Solution

```python
%%sql 


-- We create a CTE so we can clean formats a priori

WITH preCTE AS
   (  
    SELECT MIN(CAST(a.date AS date)) AS first_date, 
    MAX(CAST(a.date AS date)) AS last_date,
        COUNT(a.id) AS total_concerts,
        CONCAT (split_part(soloist_name,',',2),' ',split_part(soloist_name,',',1)) AS name
       
    FROM concerts a
       
    INNER JOIN soloists b ON a.id = b.id
       
    WHERE orchestra = 'New York Philharmonic' 
    AND soloist_instrument NOT IN ( 'Chorus', 'No Soloist',
                                 'New York Choral Artists','Schola Cantorum of NY')
    AND soloist_name NOT IN ('%choir%' ,'No Soloist')
    AND eventtype = 'Subscription Season'
       
    Group by b.soloist_name
   )

-- Corpus
SELECT name,
    to_char(first_date, 'DD Mon YYYY') AS first_date,
    to_char(last_date, 'DD Mon YYYY') AS last_date,
    total_concerts
        
FROM preCTE

ORDER BY total_concerts DESC, name ASC


-- 1% Only
LIMIT (SELECT (COUNT(name)*0.01)::INT FROM preCTE)
```
