```python
print("BENFORD LOOPS - ALGORITHIM")
print('\n')   
print("Each loops process input data according to Benford's Law")
print("Goals are: 1 Extracting stadistics data from each condition")
print("Goals are: 2 Plotting element counts data from each condition.")
print("They will be plotted after every loop.")
print("In the end a DataFrame with all stadistic data will be given")
print("Processing loops according to given input argument")

print('\n')  
print("PROCESSING STARTS")
print('\n')   


#______________________________________________________________________________________________________
# Librerias

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import kurtosis
from scipy.stats import skew
from random import shuffle

# sns style
sns.set_style('whitegrid')


#_____________________________________________________________________________________________________
# Import data


print("Extracting data...")

toneladas = pd.read_excel('IGME.xlsx')['Toneladas Metricas']

# Extraemos solo dato de inters
digits = []
for value in toneladas:
    if value == 0:
        pass
    else:
        digits.append(value)
        
print("Original data LEN (con 0s)", len(toneladas))
print("Original daa LEN (sin 0s)", len(digits))
print("\n")

# Benford column and DF creation
benford = [30.1, 17.6, 12.5, 9.7, 7.9, 6.7, 5.8, 5.1, 4.6]
benford_law = pd.DataFrame()

# Ending data DF creation
ending_data = pd.DataFrame()
mean_end = []
median_end = []
std_end = []
var_end = []
kurtosis_end = []
skew_end = []

# DF creation

toneladas = pd.DataFrame()
toneladas['digits'] = digits
print("Original data value counts: \n", toneladas.value_counts())
print('\n')

# Here NUMBER OF LOOPS IS SET
loops_interval = np.arange(1,25,4)
print("El loop is taking", loops_interval, "as % arguments")
print('\n')


#___________________________________________________________________________________________________


def simulated_errors_to_data(value):
    
    # Procedure to create and add to data random numbers (simulated errors)

    # "value" ->Sustitute for data from loop
  
    # Calculate how many number of errors according to loop condition (percent)
    try:
        number_of_errors = round(len(toneladas['digits'])/100*value, 0)
    except:
        print("VALUE argument is too high, please introduce <= 100")
        
    print("Number of errors:", number_of_errors)

    # generate data errors according to loop condition
    data_random = np.random.randint(1,9, int(number_of_errors))
    print("Random Data has been generated")

    # Calculate how many data we substracting from OG data
    eliminate_data = len(data_random)

    # Shuffle data
    toneladas['digits'] = toneladas['digits'].sample(frac=1)
    print("Data has been shuffled")

    # Reinstore data w/ errors
    reinstore = list(toneladas['digits'])[eliminate_data:]
    print("N data random =", eliminate_data, "is being added to", len(reinstore))
    
    # recreate
    toneladas['digits'] = list(data_random) +  reinstore
    print("Final digits data is LEN", len(toneladas['digits']), "Original data + ERRORS")



#_____________________________________________________________________________________________________
# Separamos DF por digitos


def digit_splitter(value):

    # Splitting DF digits
    global first_digit
    first_digit = []
    second_digit = []
    rest = []

        # STR to be able to iterate
    toneladas['digits'] = toneladas['digits'].astype('str')

        # Iterate

    for i in toneladas['digits']:
        first_digit.append(i[0])
        try:
            second_digit.append(i[1])
            rest.append(i[1:])
        except:
            second_digit.append('')
            rest.append('')
            pass
        
        # In DF union
    toneladas['first_digit'] = first_digit
    toneladas['second_digit'] = second_digit
    toneladas['rest'] = rest
    
    # OK Messages
    print("Digit splitter for value", value, "runned OK")
    print('\n')

###________________________________________________________________________________________


def zeros_dropper():
    
    # This functions drops zeros from second digit (if needed)

    # Excluimos 0 de 'second digit' y de 'resto'
    toneladas['second_digit']= toneladas['second_digit'].replace('',0)
    toneladas['second_digit']= toneladas['second_digit'].replace('0',0)

    global dos_sin_cero
    dos_sin_cero = []
    for value in toneladas['second_digit']:
        if value == 0:
            pass
        else:
            dos_sin_cero.append(value)
    

    toneladas['rest']= toneladas['rest'].replace('',0)
    toneladas['rest']= toneladas['rest'].replace('0',0)

    rest_sin_cero = []
    for value in toneladas['rest']:
        if value == 0:
            pass
        else:
            rest_sin_cero.append(value)
        
    # Some stats messages
    print('Zeros dropper has runned OK')
    print('LEN first digit', len(toneladas['first_digit']))
    print('LEN second digit', len(dos_sin_cero))
    print('LEN rest', len(rest_sin_cero))
    print('\n')



#_____________________________________________________________________________________________________
# Benford Dataframe Creation

def benford_creation():

    # This functions creates a dataframe with the proportions of BENFORD (1st column),
    #  proportion of 1st digit(2nd column), proportion of 2nd digit (3rd column)

    # Calculamos proporciones
    benford_law['Benford'] = benford
    benford_law['First Digit'] = list(toneladas['first_digit'].value_counts()/
                                      len(toneladas['first_digit'])*100)
    benford_law['Second Digit'] = list(toneladas['second_digit'].value_counts()[1:]/
                                       len(dos_sin_cero)*100)

    # Cambiamos a INT el primer valor para poder desarrollar stats
    
    global int_first
    int_first = []
    for element in first_digit:
        int_first.append(int(element))
    
    # Print Stats and Benford DF
    benford_law.index = np.arange(1,10,1)
    print(benford_law)
    print('\n')    
    
    # 
    print("Benford Creation has runned OK")
    print("LOOP",value, "STATS")
    print("1st digit mean", np.mean(int_first))
    print( "1st digit median", np.median(int_first))
    print("1st digit std", np.std(int_first))
    print( "1st digit var", np.var(int_first))
    print( "1st digit var", skew(int_first))
    print( "1st digit kurtosis", kurtosis(int_first, axis = 0, bias = False))
    
    # Append to final df data
    mean_end.append(np.mean(int_first))
    median_end.append(np.median(int_first))
    std_end.append(np.std(int_first))
    var_end.append(np.var(int_first))
    kurtosis_end.append(kurtosis(int_first, axis = 0, bias = False))
    skew_end.append(skew(int_first))

#_________________________________________________________________________________________________

def benford_plotter():

    # Benford Plooter: Plots an aggrupatted bar plot about input result

    benford_law.plot( kind = "bar", figsize=(10,8),
                     legend=['Benford','First digit'],
                     xlabel =( 'Digit (n)'),
                     ylabel = "Count of digits (n)",
                     yticks = np.arange(0,31,2),
                     title=("First & Second Digits Count Distribution for {} % of errors".format(value))
                    )

    plt.show()
    
    
#________________________________________________________________________________________________

def final_df_creation():
    
    # This functions plots ending stadistics about everyloop

    print("Creating Final DataFrame...")
    print('\n')    
    ending_data = pd.DataFrame()
    ending_data['mean'] = mean_end
    ending_data['median'] = median_end
    ending_data['var'] = var_end
    ending_data['std'] = std_end
    ending_data['skew'] = skew_end
    ending_data['kurtosis'] = kurtosis_end
    
    # Set index as loops
    ending_data.index = loops_interval
    
    # Show final DF
    print(ending_data)
    
    # Exporting final DF
    print("EXPORTING FINAL DATA...")
    ending_data.to_excel("benford_final_data.xlsx")
    print('\n')
    print("Exported as 'benford_final_data.xlsx'")
    print('\n')
    
    
#________________________________________________________________________________________________

# Run

for value in loops_interval:
    print("LOOP", value," IS RUNNING")
    print('\n')   
    simulated_errors_to_data(value)
    digit_splitter(value)
    zeros_dropper()
    benford_creation()
    benford_plotter()
    print("LOOP", value," IS FINNISHED")
    print('\n')  
    
final_df_creation()

print('PROGRAM IS FINNISHED')
```
